```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data_path = 'F:/Documents/CRCE/Project/NIDS/dataset/Edge-IIoT/Edge-IIoTset dataset/Selected dataset for ML and DL/DNN-EdgeIIoT-dataset.csv'  # Replace with your actual path
df = pd.read_csv(data_path)
df.head()
```

```python
# Separate features and label
X = df.drop(columns=["Attack_label", "Attack_type", "frame.time"])
y = df["Attack_type"]  # 0 = benign, 1 = attack

# Optional sanity check
print("Class distribution:\n", y.value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

# Columns to encode
categorical_cols = ['ip.src_host', 'ip.dst_host']

# Initialize encoders
label_encoders = {}

for col in categorical_cols:
    X[col] = X[col].astype(str)  # Ensure all values are strings
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Apply scaling only on numerical features
X_scaled = scaler.fit_transform(X)
print("Scaled Data:\n", X_scaled)  # Show first 5 rows of scaled data
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Check the shape of the split data
print("Training set shape: ", X_train.shape)
print("Testing set shape: ", X_test.shape)
```

```python
import pandas as pd

# Convert numpy arrays to pandas DataFrames/Series to reset the index
X_train = pd.DataFrame(X_train)
y_train = pd.Series(y_train)
X_test = pd.DataFrame(X_test)
y_test = pd.Series(y_test)

# Reset indices
X_train.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)

# Convert them back to numpy arrays
X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values
```

reduce train dataset

```python
# Assuming X_train and y_train are numpy arrays
reduction_ratio = 0.3  # keep 30% of data
num_samples = int(len(X_train) * reduction_ratio)

# Shuffle and select a subset
indices = np.random.permutation(len(X_train))[:num_samples]
X_train = X_train[indices]
y_train = y_train[indices]
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from torch.utils.data import Dataset, DataLoader
import numpy as np

# 1. CNN Model
class CNN(nn.Module):
    def __init__(self, num_class=15):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv1d(1, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(64, 32, kernel_size=3, padding=1)
        self.conv3 = nn.Conv1d(32, 16, kernel_size=3, padding=1)        
        self.fc1 = nn.Linear(960, 30)
        self.out_layer = nn.Linear(30, num_class)
        self.relu = nn.ReLU()

    def forward(self, x):
        # Conv layer 1
        x = self.conv1(x)
        x = F.relu(x)
        
        # Conv layer 2
        x = self.conv2(x)
        x = F.relu(x)
        
        # Conv layer 3
        x = self.conv3(x)
        x = F.relu(x)
        
        # Max pooling
        x = F.max_pool1d(x, kernel_size=1)
        
        # Print shape after conv layers
        # print("Shape before flattening: ", x.shape)
    
        # Flatten layer 
        x = x.reshape(x.size(0), -1)
    
        # FC layer 1
        x = self.fc1(x)
        x = F.relu(x)
        
        # Output Layer 
        x = self.out_layer(x)
        output = F.log_softmax(x, dim=1)
        return output


# 2. Create Dataset Class
class NIDS_Dataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        # Convert X to tensor of shape (1, num_features) and y to tensor
        sample = torch.tensor(self.X[idx], dtype=torch.float).unsqueeze(0)  # unsqueeze to add channel dimension
        label = torch.tensor(self.y[idx], dtype=torch.long)
        return sample, label

from torch.utils.data import DataLoader

# 3. Prepare Data Loaders
train_dataset = NIDS_Dataset(X_train, y_train)
test_dataset = NIDS_Dataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)


# 4. Initialize Model, Loss Function, and Optimizer
model = CNN(num_class=15)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs=10

# 5. Train the Model
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch_idx, (inputs, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

        print(f"Epoch {epoch+1}/{num_epochs}, Batch {batch_idx+1}/{len(train_loader)}", end='\r')

    print(f"\nEpoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader):.4f}")

# 6. Evaluate the Model
model.eval()
y_true = []
y_pred = []

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        
        y_true.extend(labels.numpy())
        y_pred.extend(predicted.numpy())

# Evaluate performance
accuracy = accuracy_score(y_true, y_pred)
print("Accuracy on Test Data: ", accuracy)

# Display classification report for detailed evaluation
print("\nClassification Report:\n", classification_report(y_true, y_pred))

# Display confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_true, y_pred))
```
