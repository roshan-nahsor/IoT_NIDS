```python
import numpy as np
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler

# Step 1: Load the new CSV file (replace with your actual file path)
input_csv = '../live_dataset_1.csv'  # Replace with your file path
df = pd.read_csv(input_csv)
```

```python
try:
    df['frame.time'] = pd.to_datetime(df['frame.time'], infer_datetime_format=True)
    time_difference = df['frame.time'].diff().fillna(pd.Timedelta(seconds=0))
    df.insert(1, 'time_difference', time_difference)
    print(df.head())
except ValueError as e:
    print(f"Error with infer_datetime_format: {e}")
```

```python
drop_columns = ["frame.time", "ip.src_host", "ip.dst_host", "arp.src.proto_ipv4", "arp.dst.proto_ipv4", 
                 "http.file_data", "http.request.full_uri", "icmp.transmit_timestamp",
                 "http.request.uri.query", "tcp.options", "tcp.payload", "tcp.srcport",
                 "tcp.dstport", "udp.port", "mqtt.msg", 'mbtcp.len', 'mbtcp.trans_id', 'mbtcp.unit_id', 'dns.qry.name.len',
               'icmp.unused', 'http.tls_port', 'dns.qry.type', 'mqtt.msg_decoded_as']

df_dropped=df.drop(drop_columns, axis=1, inplace=False)
df_dropped.head()
```

```python
categorical_columns = [
    'http.request.method',
    'http.referer',
    'http.request.version',
    'http.response',
    'dns.qry.name',
    'mqtt.protoname',
    'mqtt.topic',
    'mqtt.conack.flags'
]

from sklearn.preprocessing import LabelEncoder

for col in categorical_columns:
    df[col] = LabelEncoder().fit_transform(df_dropped[col].astype(str))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_dropped['tcp.checksum'].dtype
```

```python
# Convert the 'hex_column' to decimal
df_dropped['tcp.checksum'] = df_dropped['tcp.checksum'].apply(lambda x: int(x, 16))

# If you want to overwrite the original column:
# df_dropped['hex_column'] = pd.to_numeric(df_dropped['hex_column'], base=16)

# Print the DataFrame to see the result
print(df_dropped)

# Print the data type of the new (or converted) column
print(df_dropped['tcp.checksum'].dtype)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in df_dropped.columns:
    if df_dropped[col].dtype == 'object':
        le = LabelEncoder()
        df_dropped[col] = le.fit_transform(df_dropped[col])
```

```python
numerical_cols = df_dropped.columns[df_dropped.nunique() > 2]  # Likely numeric
df_dropped[numerical_cols] = scaler.fit_transform(df_dropped[numerical_cols])
```

```python
df_dropped.shape[0]
```

```python

# Step 3: Load the saved model (weights and bias)
with open('pso_optimized_logistic_regression_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Extract weights and bias from the loaded model
weights = model['weights']
bias = model['bias']

# Step 4: Perform inference (logistic regression prediction)
logits = df_dropped @ weights + bias  # Compute logits (linear combination of inputs and weights)
probs = 1 / (1 + np.exp(-logits))  # Apply sigmoid to get probabilities

# Step 5: Convert probabilities to binary predictions (0 or 1)
y_pred = (probs >= 0.5).astype(int)  # Thresholding at 0.5 for binary classification

# Step 6: Output the predictions
df_dropped['Prediction'] = y_pred  # Add predictions to the original data
df_dropped.to_csv('predictions.csv', index=False)  # Save predictions to a new CSV file

print("Inference complete. Predictions saved to 'predictions.csv'.")
```
